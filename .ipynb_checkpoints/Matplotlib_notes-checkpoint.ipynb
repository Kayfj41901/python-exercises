Basic Graph

```python
x = [0,1,2,3,4,]
y = [0,2,4,6,8]
#resize your Graph 
plt.figure(figsize=(5,3), dpi=150)

# Line 1
# Keyword Argument Notation 
# Shorthand notation 
#fmt = '[color][marker][line]'
# you could condense all of the extra formatting to something like 'r^--'
plt.plot(x,y, label='2x', color='red', linestyle= '--', linewidth=2, markersize=10, marker='.')

##line number 2
#select interval we want to plot points at 
x2 = np.arange(0,4,0.5)

#plot graph
plt.plot(x2, x2**2, label='x^2')

#Add a title 
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

#add a y and x label 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#x and y axis tickmarks 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#add a legend, add a third space to plt.plot with name of legend 
plt.legend()

plt.savefig('mygraph.png', dpi=300)


plt.show()
```

Bar Chart 

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.bar(labels, values)

plt.figure(figsize=(6,4))

plt.show()
```
